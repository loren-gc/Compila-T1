## Escrevendo um Analisador Léxico
O processo de “análise léxica” consiste em ler um texto de entrada e dividí-lo em um fluxo de tokens. Cada token é como uma palavra válida do dicionário. Essencialmente, o papel do analisador léxico é simplesmente certificar-se de que o texto de entrada consiste em símbolos e tokens válidos antes de qualquer processamento adicional relacionado à análise.

Cada token é definido por uma expressão regular. Assim, sua tarefa aqui é definir um conjunto de expressões regulares para a linguagem uZig. O trabalho real de análise léxica será tratado pelo SLY. Para um melhor entendimento, estude o capítulo [“Escrevendo um Analisador Léxico”](https://sly.readthedocs.io/en/latest/sly.html#writing-a-lexer)
da documentação do SLY.

### Especificação
Seu analisador léxico deve reconhecer os símbolos e tokens da gramática do uZig. Por exemplo, no exemplo abaixo, o nome à esquerda é o nome do token e o valor à direita é o texto correspondente:

Palavras-chave Reservadas:
```
    KEYWORD_while : 'while'
    KEYWORD_if    : 'if'
    KEYWORD_else  : 'else'
```

Identificadores:
```
    IDENTIFIER    : qualquer texto que comece com uma letra ou '_', seguido por qualquer número
                    de letras, dígitos ou sublinhados, que não seja uma palavra reservada
```

Alguns Operadores e Delimitadores:
```
    PLUS          : '+'
    MINUS         : '-'
    ASTERISK      : '*'
    SLASH         : '/'
    EQUAL         : '='
    SEMI          : ';'
    LPAREN        : '('
    RPAREN        : ')'
```

Literais:
```
    INTEGER       : 123
    STRINGLITERAL : "Ola Mundo\n"
```

Para `INTEGER`, você deve considerar apenas números decimais.


Comentários: Para ser ignorado pelo seu analisador léxico
```
     //             Ignora o resto da linha
```

Erros: seu analisador léxico deve relatar as seguintes mensagens de erro:
```
     <lineno>: Illegal character '<value>'
     <lineno>: Unterminated character constant
     <lineno>: Unterminated string literal
```

## Anexo
A lista completa de tokens usados na linguagem uZip é:

* Palavras-reservadas:

```
    KEYWORD_and          : 'and'
    KEYWORD_break        : 'break'
    KEYWORD_const        : 'const'
    KEYWORD_continue     : 'continue'
    KEYWORD_else         : 'else'
    KEYWORD_false        : 'false'
    KEYWORD_if           : 'if'
    KEYWORD_or           : 'or'
    KEYWORD_true         : 'true'
    KEYWORD_var          : 'var'
    KEYWORD_while        : 'while'
```

* Identificadores:

```
    IDENTIFIER           : qualquer texto que comece com uma letra ou '_', seguido por qualquer número
                           de letras, dígitos ou sublinhados, que não seja uma palavra reservada
    BUILTINIDENTIFIER    : qualquer texto que comece com '@', seguido por uma letra ou '_', e então
                           seguido por qualquer número de letras, dígitos ou sublinhados
```

* Constantes:

```
    CHAR_LITERAL         : qualquer texto delimitado por aspas simples
    FLOAT                : qualquer texto que contenha um '.' e um ou mais dígitos de 0 a 9, podendo ter
                           dígitos antes e depois do ponto, apenas antes do ponto, ou apenas depois do ponto
    INTEGER              : qualquer texto composto por um ou mais dígitos de 0 a 9
    STRINGLITERAL        : qualquer texto delimitado por aspas duplas
```

* Operadores:

```                    
    PLUS                 : '+'
    MINUS                : '-'
    ASTERISK             : '*'
    SLASH                : '/'
    PERCENT              : '%'
    EQUALEQUAL           : '=='
    EXCLAMATIONMARKEQUAL : '!='
    LARROWEQUAL          : '<='
    LARROW               : '<'
    RARROWEQUAL          : '>='
    RARROW               : '>'
    EXCLAMATIONMARK      : '!'
```

* Atribuição:

```    
    EQUAL             : '='
```

* Delimitadores:

```
    COLON             : ':'
    LPAREN            : '('
    RPAREN            : ')'    
    LBRACE            : '{'
    RBRACE            : '}'    
    COMMA             : ','
    SEMI              : ';'
```

A linguagem uZig suporta apenas comentários de uma única linha:

```
     //             Ignora o resto da linha
```

**Observação:** A linguagem uZig não suporta comentários de múltiplas linhas (como `/* ... */` do C). Para comentar várias linhas, você deve prefixar cada linha com `//`.
